```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import json
import os
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
with open ("crix.json") as f:
    data = json.load(f)
df = pd.DataFrame.from_dict(data,orient='columns')
df
```

```python
cluster_data = make_blobs(n_samples=200,n_features=2,centers=5,cluster_std=1.6,random_state=50)
points = cluster_data[0]
plt.scatter(cluster_data[0][:,0],cluster_data[0][:,1],c=cluster_data[1],cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
```

```python
# K means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(points)
print(kmeans.cluster_centers_)
y_km = kmeans.fit_predict(points)
```

```python
plt.scatter(points[y_km==0,0],points[y_km==0,1],s=100,c='red')
plt.scatter(points[y_km==1,0],points[y_km==1,1],s=100,c='orange')
plt.scatter(points[y_km==2,0],points[y_km==2,1],s=100,c='yellow')
plt.scatter(points[y_km==3,0],points[y_km==3,1],s=100,c='green')
plt.scatter(points[y_km==4,0],points[y_km==4,1],s=100,c='blue')
```

```python
# Hierarchical Clusterinng
dendrogram = sch.dendrogram(sch.linkage(points,method="ward"))
hc = AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
y_hc = hc.fit_predict(points)
```
