```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# There are seven sub-sectors under fund news in sina website
url_id = {'动态':'56941',
          '看市':'56934',
          '评论':'56942',
          '持仓':'56945',
          '研报':'56946',
          '业界':'56948',
          '人物':'56952'}

url_dict={}
for k,v in url_id.items():
    url_dict[k]="https://finance.sina.com.cn/roll/index.d.html?cid="+v+"&page="
print(url_dict)
```

```python
# Initialize the header and construct a dataframe for storing scraped data
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
df = pd.DataFrame({"category":"","time":"","title":"","href":"","content":""},index=["0"])
```

```python
# define a function for getting the content of news
def get_content(url):
    res = requests.get(url,headers = headers)
    res.encoding = "utf-8"
    soup = BeautifulSoup(res.text,"html.parser")
    target = soup.select('div[class="article"] p')
    content = ""
    for i in range(len(target)-1):
        content+=target[i].text
    return content
```

```python
# scrape the title, corresponding url, publishing time of these seven categories
for k,v in url_dict.items():
    print(v)
    for i in range(20):
        print(i)
        url = v+str(i+1)
        res = requests.get(url,headers = headers)
        res.encoding = "utf-8"
        soup = BeautifulSoup(res.text,"html.parser")
        target = soup.select("ul li")
        target = target[2:]
        for each in target:
            temp = {}
            temp["category"]= k
            temp["href"] = each.a["href"]
            temp["title"] = each.a.text
            temp["time"] = each.span.text[1:-1]
            print(each.a["href"])
            temp["content"] = get_content(each.a["href"])
            #print(temp)
            df = df.append(temp,ignore_index = True)
```

```python
# It will take a long long time, just don't need to run it again
# drop the first row which is empty and display the result
df = df.drop([0])
df
```

```python
# save the data to local
df.to_csv("sina_fund_news.csv",encoding ="utf_8_sig")
```
